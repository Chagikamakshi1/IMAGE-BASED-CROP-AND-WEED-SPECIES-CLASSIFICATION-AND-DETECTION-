```python
from PIL import Image
import numpy as np
import os
dst_img = r"D:\final images\maize segment\HSV maize"
#listing files in images folder
list_img = os.listdir(dst_img)
#iterating over dst_image to get the images as arrays
for image in sorted(list_img):
    [file_name, ext] = os.path.splitext(image) #splitting file name from its extension
    
    arr = np.array(Image.open(os.path.join(dst_img, image))) #creating arrays for all the images
    [h, w] = np.shape(arr)[0:2]#calculating height and width for each image
    arr_dim = arr.ndim #calculating the dimension for each array
    arr_shape = arr.shape #calculating the shape for each array
    if arr_dim == 2:
        arr_mean = np.mean(arr)
        print(f'{file_name}, greyscale={arr_mean:.1f}')
    else:
        arr_mean = np.mean(arr, axis=(0,1))
        if len(arr_mean) == 3: #RGB CASE
            print(f'{file_name}, R={arr_mean[0]:.1f},  G={arr_mean[1]:.1f}, B={arr_mean[2]:.1f} ')
        else: #ALPHA CASE
            print(f'{file_name}, R={arr_mean[0]:.1f}, G={arr_mean[1]:.1f}, B={arr_mean[2]:.1f}, ALPHA={arr_mean[3]:.1f}')
```

# THIS IS FINAL MASKED IMAGE AFTER ORIGINAL--->LAB--->THRESHOLD--->MASK--->FILTER(DENOISE)

```python
# read image in BGR
import os
import cv2
f=r"D:\resize\resize peanut"
img_number=1
img_number1=1
for file in os.listdir(f):
    f_img=f+"/"+file

    img = cv2.imread(f_img)
# convert to LAB space
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
#     cv2.imwrite(r"D:\final images\okra segment\simply lab\maize"+str(img_number)+".jpg",lab)
#     img_number+=1
# store the a-channel
    a_channel = lab[:,:,1]
# Automate threshold using Otsu method
    th = cv2.threshold(a_channel,127,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1]
    cv2.imwrite(r"D:\final images\peanut segment\llab ostu peanut\lab otsu peanut"+str(img_number)+".jpg",th)
    img_number+=1
# Mask the result with the original image
    masked = cv2.bitwise_and(img, img, mask = th)
    bilateral_filter = cv2.bilateralFilter(src=masked, d=9, sigmaColor=75, sigmaSpace=75)
    cv2.imwrite(r"D:\final images\peanut segment\lab mask threshold filtered peanut\final peanut"+str(img_number1)+".jpg",bilateral_filter)
    img_number1+=1
```

# Apply Bilateral Filtering

```python
# """
# Apply Bilateral Filtering
# """
# Using the function bilateralFilter() where d is diameter of each...
# ...pixel neighborhood that is used during filtering.
# sigmaColor is used to filter sigma in the color space.
# sigmaSpace is used to filter sigma in the coordinate space.
import cv2
import numpy as np
f=r"D:\resize\resize maize"
img_number=1
for file in os.listdir(f):
    f_img=f+"/"+file
    image = cv2.imread(r"D:\resize real.png")
    bilateral_filter = cv2.bilateralFilter(src=image, d=9, sigmaColor=75, sigmaSpace=75)

    cv2.imshow('Original', image)
    cv2.imshow('D:/Bil Filtering', bilateral_filter)

    cv2.waitKey(0)
    cv2.imwrite('D:/bilateral_filtering.jpg', bilateral_filter)
    cv2.destroyAllWindows()
```
