```python
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"D:\resize\resize okra\resize okra 1 2 .png")
```

```python
hsv_img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
h,s,v=hsv_img[:,:,0],hsv_img[:,:,1],hsv_img[:,:,2]
```

```python
clahe=cv2.createCLAHE(clipLimit=5.0,tileGridSize=(8,8))
v=clahe.apply(v)
hsv_img=np.dstack((h,s,v))
rgb=cv2.cvtColor(hsv_img,cv2.COLOR_HSV2RGB)
plt.imshow(rgb);
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image=cv2.imread(r"D:\resize_images\peanut_resize_6-9-22 1 .png")
image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)  
# cv2.imshow("1",image)
```

```python
hist=cv2.calcHist([image],[0],None,[256],[0,256]) #image.mask,histsize(no of bins i.e equal to number of pixels values,range
plt.plot(hist)

```

```python
# cv2.imshow("image",image) here we can see  all values are confined between so equalize
image_hist=cv2.equalizeHist(image)

hist=cv2.calcHist([image_hist],[0],None,[256],[0,256]) #image.mask,histsize(no of bins i.e equal to number of pixels values,range
plt.plot(hist)
plt.show()
cv2.imshow("image",image)
cv2.imshow("applied histogram",image_hist)
cv2.waitKey(0)
```
